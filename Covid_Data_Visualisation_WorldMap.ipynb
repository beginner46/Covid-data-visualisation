```python
#importing the necessary libraries

import pandas as pd
import pycountry
import matplotlib.pyplot as plt
import numpy as np
import random 
from covid import Covid
from matplotlib.ticker import AutoMinorLocator
from matplotlib import gridspec
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

```python
#using the john hopkins data set

covid =Covid()
df = pd.DataFrame(covid.get_data())
df.head() #getting the data frame
```

```python
df.shape
```

```python
df.head()
```

```python
df_countries = df.iloc[8:, :]
```

```python
df_countries
```

```python
def to_code(country):
    try:
        result = pycountry.countries.search_fuzzy(country)
        return result[0].alpha_3
    except Exception:
        return np.nan

df_countries["country_code"] = df_countries["country"].apply(lambda country: to_code(country))
df_countries
```

```python
#No. of confirmed cases

world_map = px.choropleth(df_countries,
                          locations='country_code',
                          color='confirmed', 
                          hover_name='country', 
                          color_continuous_scale=px.colors.sequential.Aggrnyl_r)
 
world_map.update_layout(coloraxis_showscale=True,)
 
world_map.show()
```

```python
#no. of deaths

world_map = px.choropleth(df_countries,
                          locations='country_code',
                          color='deaths', 
                          hover_name='country', 
                          color_continuous_scale=px.colors.sequential.Aggrnyl_r)
 
world_map.update_layout(coloraxis_showscale=True,)
 
world_map.show()
```

```python
#using the world-o-meter data set

covid =Covid(source = "worldometers")
df1 = pd.DataFrame(covid.get_data())
df1.head() #getting the data frame
```

```python
df_countries1 = df1.iloc[8:, :]
```

```python
df_countries1
```

```python
def to_code(country):
    try:
        result = pycountry.countries.search_fuzzy(country)
        return result[0].alpha_3
    except Exception:
        return np.nan

df_countries1["country_code"] = df_countries1["country"].apply(lambda country: to_code(country))
df_countries1
```

```python
#no. of deaths

for c in df_countries1.columns:
  world_map = px.choropleth(df_countries1,
                          locations='country_code',
                          color = c, 
                          hover_name='country', 
                          color_continuous_scale=px.colors.sequential.Aggrnyl_r)
 
  world_map.update_layout(coloraxis_showscale=True,)
 
  world_map.show()
  print(f"showing the {c} over a world map")
  print(end = "\n") 
  
```
