```python
#plotting the values of confirmed cases corresponding to the first few countries similar could be extended for any number of countries

#barplot

df_initial.plot.bar(x = 'country', y = 'confirmed')
plt.grid()

#scatterplot

print(end = "\n")
np.random.seed(seed = 30)
cmap = matplotlib.cm.get_cmap('Spectral')
col_dict1 = pd.Series({k:cmap(np.random.rand()) for k in df_initial['confirmed'].unique()})
col_dict1.name = 'col_dict1'
df_initial = pd.merge(df_initial, col_dict1, how = 'left', left_on = 'confirmed', right_index = True)
df_initial.plot.scatter(x = 'country', y = 'confirmed', s = (df_initial['confirmed']/100000), c = df_initial['col_dict1']);
plt.xticks(rotation = 90)
plt.grid()
plt.show()

#line plot

print(end = "\n")
df_initial.plot(kind = 'line', x='country', y='confirmed', color='blue')
plt.xticks(rotation = 90)
plt.grid()
plt.show()

#pie chart


data = df_initial['confirmed']

print("\n")
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
countries = df_initial['country']
ax.pie(data, labels = countries, autopct='%1.2f%%')
plt.legend(loc=2, fontsize = 'x-small')
plt.show()
  
```

```python
#plotting the values of total deaths corresponding to the first few countries similar could be extended for any number of countries

#barplot


df_initial.plot.bar(x = 'country', y = 'deaths', color = 'red')
plt.grid()
print(end = "\n")

#scatterplot

print(end = "\n")
np.random.seed(seed = 30)
cmap = matplotlib.cm.get_cmap('Spectral')
col_dict2 = pd.Series({k:cmap(np.random.rand()) for k in (df_initial.get('deaths')).unique()})
col_dict2.name = 'col_dict2'
df_initial = pd.merge(df_initial, col_dict2, how = 'left', left_on = 'deaths', right_index = True)
df_initial.plot.scatter(x = 'country', y = 'deaths', s = (df_initial['deaths']/1000), c = df_initial['col_dict2']);
plt.xticks(rotation = 90)
plt.grid()
plt.show()

#line plot

print(end = "\n")
df_initial.plot(kind = 'line', x='country',y='deaths',color='red')
plt.xticks(rotation = 90)
plt.legend()
plt.grid()
plt.show()

#pie chart
data1 = df_initial['deaths']

print("\n")
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
countries = df_initial['country']
ax.pie(data1, labels = countries, autopct='%1.2f%%')
plt.legend(loc=2, fontsize = 'x-small')
plt.show()
```

```python
#using the world-o-meter data set

covid =Covid(source = "worldometers")
df1 = pd.DataFrame(covid.get_data())
df1.head() #getting the data frame
```

```python
df_countries1 = df1.iloc[8:, :] #removing values for the continent
```

```python
df_countries1
```

```python
def to_code(country):
    try:
        result = pycountry.countries.search_fuzzy(country)
        return result[0].alpha_3
    except Exception:
        return np.nan

df_countries1["country_code"] = df_countries1["country"].apply(lambda country: to_code(country))
df_countries1
```

```python
#plotting it on the world map

for c in df_countries1.columns:
  world_map = px.choropleth(df_countries1,
                          locations='country_code',
                          color = c, 
                          hover_name='country', 
                          color_continuous_scale=px.colors.sequential.Aggrnyl_r)
 
  world_map.update_layout(coloraxis_showscale=True,)
 
  world_map.show()
  print(f"showing the {c} over a world map")
  print(end = "\n") 
```
